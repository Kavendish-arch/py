```python
img = cv.imread(img1)

def bgr2gray_ave(img):
    h, w, l = img.shape[:]
    gray_img = np.zeros((h,w), dtype=np.uint8)
    for row in range(h):
        for col in range(w):
            gray_img[row, col] = ave(img[row, col])
    plt.imshow(gray_img)
    plt.show()

def ave(num):
    return np.sum(num) / len(num)
ave(np.array([1,2,3]))
```

```python
img = cv.imread(img1)

def bgr2gray_max(img):
    h, w, l = img.shape[:]
    gray_img = np.zeros((h,w), dtype=np.uint8)
    for row in range(h):
        for col in range(w):
            gray_img[row, col] = np.max(img[row, col])
    plt.imshow(gray_img)
    plt.show()
bgr2gray_max(img)
```

```python
img = cv.imread(img1)

def bgr2gray_min(img):
    h, w, l = img.shape[:]
    gray_img = np.zeros((h,w), dtype=np.uint8)
    for row in range(h):
        for col in range(w):
            gray_img[row, col] = np.min(img[row, col])
    plt.imshow(gray_img)
    plt.show()
bgr2gray_min(img)
```
