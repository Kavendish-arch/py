# Pandas - Series

结构化数据分析利器（依赖numpy）

提供了多种高级数据结构

拥有大数据索引和处理能力


Pandas中的主要数据对象是Series和DataFrame。

## Series
* 创建series
* 改变dtype类型
* Series性质
* 通过index排序
* 数据选择

Series类似一维数组对象，包含一个数组数据和一个与数组关联的数据标签，被叫做索引 

```python
Image(filename='series.png')
```

## 创建 Series
* 通过list
* 通过dict

### 通过list

```python
pd.Series(['a','b','c','d','e','f','g'])
```

```python
pd.Series(['a','b','c','d','e','f','g'],index=[1,2,3,4,5,6,7])
```

```python
list(range(1,8))
```

```python
pd.Series(['a','b','c','d','e','f','g'],index=list(range(1,8)))
```

```python
pd.Series(['a','b','c','d','e','f','g'],index=list(range(1,8)),name='NAME')
```

```python
pd.Series([1,2,3,4],dtype='str')
```

### 通过dict

```python
pd.Series({1:'a',2:'b',3:'c',4:'d',5:'e',6:'f',7:'g'},name='NAME')
```

## 通过index排序

```python
s=pd.Series(['a','b','c','d'],index=[5,1,3,8])
s
```

```python
s.sort_index()
```

```python
s.sort_index(ascending=False)
```

```python
s
```

```python
s.sort_index(inplace=True)
```

```python
s
```

```python
s.sort_values()
```

```python
s.sort_values(ascending=False)
```

```python
s
```

```python
s.sort_values(inplace=True)
s
```

## Series选择
* 通过位置选择
* 通过index选择
* 通过条件选择

### 通过位置选择

```python
s
```

```python
s.iloc[0]
```

```python
s.iloc[:3]
```

```python
s.iloc[[0,2,3]]
```

```python
s.iloc[[0,-2,-1]]
```

### 通过index选择

```python
s
```

```python
s.loc[5]
```

```python
s.loc[[5,8]]
```

```python
s[5]
```

```python
s[[5,8]]
```

### 通过条件选择

```python
s
```

```python
s>'b'
```

```python
s[s>'b']
```

```python
s.loc[s>'b']
```

```python
(s>'b')& (s<'d')
```

```python
s.loc[(s>'b')& (s<'d')]
```

```python
s.gt('b')& s.lt('d')
```

```python
s.loc[s.gt('b')& s.lt('d')]
```
