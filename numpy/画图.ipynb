```python
x = np.linspace(0, 10, 20)
y = np.sin(x)

plt.plot(x, y, label='sin(x)', markersize=10)
plt.plot(x, np.cos(x), 'o',label='cos(x)', markersize=5)
plt.legend(loc='upper right')
#plt.pegendplt.legend?
```
